## Armstrong Number

```python
num = input('Enter Armstrong Number :')
temp = 0

for i in num:
    temp = temp + int(i)**3

x = str(temp)

if x == num:
    print('The number is Armstrong')

elif x != num:
    print('Not Armstrong')

else:
    print('Error')
```

# simple calculator

```python
num1 = int(input('Enter 1st number : '))
num2 = int(input('Enter 2nd number : '))

sing = input('Enter Operator +, -, * or / : \n')

if sing == "+":
    print('Result : ',num1+num2)

elif sing == "-":
    print('Result : ',num1-num2)

elif sing == "*":
    print('Result : ',num1*num2)

elif sing == "/":
    print('Result : ',num1/num2)

else:
    print('Error Feild')

```
