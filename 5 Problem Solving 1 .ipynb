# 6 Armstrong Number

```python
num = input('Enter Armstrong Number :')
temp = 0

for i in num:
    temp = temp + int(i)**3

x = str(temp)

if x == num:
    print('The number is Armstrong')

elif x != num:
    print('Not Armstrong')

else:
    print('Error')
```

# 8 Febonacci

```python
nterms = int(input("How many terms? "))

n1 = 0
n2 = 1
count = 0

if nterms <= 0:
    print("please enter a positive integer")

elif nterms == 1:
    print("Febonacci sequence upto:", nterms, ":")
    print(n1)
else:
    print('Febonacci sequence:')
    while count < nterms:
        print(n1)
        nth = n1 + n2
        #update values
        n1 = n2
        n2 = nth
        count += 1
```

```python
# Reverse the sequence of the list.
```

```python
list1 = []
count = 0

for i in range(2,101):
    if i%2 == 0:
        list1.append(i)
        count += 1


list1.reverse()

print(list1)
```

```python
# Sort the list in ascending and descending order.
```

```python
list1 = []
count = 0

for i in range(2,101):
    if i%2 == 0:
        list1.append(i)
        count += 1


list1.sort()
print('Ascending : ',list1)
print("---------------------------------------")
list1.sort(reverse=True)
print('Descending :', list1)
```
